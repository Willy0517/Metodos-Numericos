```python
import math

print("=== MÉTODO DE RUNGE-KUTTA ORDEN 4 (RK4) ===\n")
print("Ingrese la función f(x,y) usando 'x' y 'y'. Ejemplo: x+y,  x**2 - y,  y-x\n")

func_str = input("f(x,y) = ")

def f(x, y):
    return eval(func_str, {"x": x, "y": y, "math": math})

x0 = float(input("Ingrese x0: "))
y0 = float(input("Ingrese y0: "))
xf = float(input("Ingrese xf: "))
h = float(input("Ingrese el paso h: "))

print("\n=== RESULTADOS ===")

x = x0
y = y0

while x <= xf + 1e-9:
    print(f"x = {x:.4f}, y = {y:.6f}")

    k1 = f(x, y)
    k2 = f(x + h/2, y + (h/2)*k1)
    k3 = f(x + h/2, y + (h/2)*k2)
    k4 = f(x + h,   y + h*k3)

    y = y + (h/6)*(k1 + 2*k2 + 2*k3 + k4)
    x = x + h

```
