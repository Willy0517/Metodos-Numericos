```python
import math

def fixed_point_iteration(g, x0, tol, max_iter):
    """
    Implementa el método de punto fijo para encontrar una raíz de una función.

    Args:
        g: La función g(x) = x.
        x0: La aproximación inicial.
        tol: La tolerancia para el criterio de parada.
        max_iter: El número máximo de iteraciones.

    Returns:
        La aproximación de la raíz si se encuentra dentro del número máximo de iteraciones,
        de lo contrario, None.
    """
    x = x0
    for i in range(max_iter):
        x_next = g(x)
        if abs(x_next - x) < tol:
            print(f"Convergencia alcanzada en la iteración {i+1}")
            return x_next
        x = x_next
        print(f"Iteración {i+1}: x = {x}")
    print("El método no convergió dentro del número máximo de iteraciones.")
    return None

# Solicitar entrada de datos al usuario
function_str = input("Introduce la función g(x) en términos de x (ej: math.cos(x) + x - 1): ")
x0_input = float(input("Introduce la aproximación inicial x0: "))
tol_input = float(input("Introduce la tolerancia tol: "))
max_iter_input = int(input("Introduce el número máximo de iteraciones: "))

# Crear la función g(x) a partir de la cadena de entrada
# Es importante ser cauteloso al usar eval() con entrada de usuario
# Para este ejemplo educativo, se asume que la entrada es segura.
# En aplicaciones de producción, considera alternativas más seguras.
g_func = lambda x: eval(function_str)

# Ejecutar el método de punto fijo
root = fixed_point_iteration(g_func, x0_input, tol_input, max_iter_input)

if root is not None:
    print(f"\nLa raíz aproximada es: {root}")
```
