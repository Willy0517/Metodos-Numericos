```python
# Obtener datos del usuario
x_data_input = input("Introduce los valores de x separados por comas (ej: 1.0, 2.0, 3.0): ")
y_data_input = input("Introduce los valores de y separados por comas (ej: 1.0, 4.0, 9.0): ")
x_interpolate_input = input("Introduce el valor de x donde deseas interpolar (ej: 2.5): ")

# Convertir los datos de entrada a listas de floats
try:
    x_data_manual = [float(x.strip()) for x in x_data_input.split(',')]
    y_data_manual = [float(y.strip()) for y in y_data_input.split(',')]
    x_interpolate_manual = float(x_interpolate_input.strip())

    # Verificar si el número de puntos x e y coincide
    if len(x_data_manual) != len(y_data_manual):
        print("Error: El número de valores x e y no coincide.")
    else:
        # Realizar la interpolación con los datos ingresados manualmente
        result_manual = neville_interpolation(x_data_manual, y_data_manual, x_interpolate_manual)
        print(f"El valor interpolado en {x_interpolate_manual} es: {result_manual}")

except ValueError:
    print("Error: Asegúrate de ingresar números válidos separados por comas.")
except Exception as e:
    print(f"Ocurrió un error: {e}")
```
