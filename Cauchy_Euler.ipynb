```python
import math

print("===  (Euler Mejorado / Heun) ===\n")
print("Ingrese la funcion f(x,y) usando 'x' y 'y'. Ej: x+y, x**2-y, etc.\n")

func_str = input("f(x,y) = ")

# Convertir string a funci√≥n
def f(x, y):
    return eval(func_str)

# Datos iniciales
x0 = float(input("x0 = "))
y0 = float(input("y0 = "))
xf = float(input("xf = "))
h = float(input("h = "))

print("\n=== RESULTADOS ===")
x = x0
y = y0

while x <= xf + 1e-12:
    print(f"x = {x:.4f}, y = {y:.6f}")

    # Predictor (Euler simple)
    y_pred = y + h * f(x, y)

    # Corrector (promedio de pendientes)
    y = y + (h/2) * (f(x, y) + f(x + h, y_pred))

    x = x + h


```
