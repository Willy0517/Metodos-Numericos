```python
import numpy as np

def lagrange_interpolation(x_values, y_values, x):
  """
  Implementa el método de interpolación de Lagrange.

  Args:
    x_values: Una lista o array de valores x conocidos.
    y_values: Una lista o array de valores y conocidos correspondientes a x_values.
    x: El valor en el que se desea interpolar.

  Returns:
    El valor interpolado en x.
  """
  n = len(x_values)
  if n != len(y_values):
    raise ValueError("Los arrays de x_values y y_values deben tener la misma longitud.")

  result = 0
  for i in range(n):
    term = y_values[i]
    for j in range(n):
      if i != j:
        term *= (x - x_values[j]) / (x_values[i] - x_values[j])
    result += term
  return result

# Entrada de datos por el usuario
try:
    x_conocidos_str = input("Introduce los valores de x conocidos separados por comas (ej: 0,1,2,3): ")
    y_conocidos_str = input("Introduce los valores de y conocidos separados por comas (ej: 1,2,4,8): ")
    x_a_interpolar_str = input("Introduce el valor de x a interpolar (ej: 1.5): ")

    x_conocidos = np.array([float(x) for x in x_conocidos_str.split(',')])
    y_conocidos = np.array([float(y) for y in y_conocidos_str.split(',')])
    x_a_interpolar = float(x_a_interpolar_str)

    y_interpolado = lagrange_interpolation(x_conocidos, y_conocidos, x_a_interpolar)
    print(f"El valor interpolado en x = {x_a_interpolar} es y = {y_interpolado}")

except ValueError as e:
    print(f"Error en la entrada de datos: {e}")
    print("Por favor, asegúrate de introducir números separados por comas y un número para el punto a interpolar.")
```
