```python
import math

import math

# --------------------------------------------
# Evalúa f(x, y) escrita como texto
# --------------------------------------------
def f_eval(f_str, x, y):
    return eval(f_str, {"x": x, "y": y, "math": math, **math.__dict__})

# --------------------------------------------
# Método de Euler con h ingresado
# --------------------------------------------
def euler_h(f_str, x0, y0, xf, h):
    n = int((xf - x0) / h)  # número de pasos
    x, y = x0, y0

    print("\nResultados (Método de Euler):")
    print("i\t x\t\t y")
    print("--------------------------------")

    for i in range(n + 1):
        print(f"{i}\t {x:.5f}\t {y:.5f}")
        y = y + h * f_eval(f_str, x, y)   # fórmula de Euler
        x = x + h

# --------------------------------------------
# PROGRAMA PRINCIPAL
# --------------------------------------------
print("\n=== MÉTODO DE EULER – CON h INGRESADO ===\n")

f_str = input("Ingrese f(x,y) (ej: x + y): ")
x0 = float(input("Ingrese x0: "))
y0 = float(input("Ingrese y0: "))
xf = float(input("Ingrese el valor final xf: "))
h = float(input("Ingrese el valor del paso h: "))

euler_h(f_str, x0, y0, xf, h)

```
