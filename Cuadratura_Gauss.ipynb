```python
import numpy as np
import math

# ---------------------------------------------
# Evalúa una función escrita como texto
# ---------------------------------------------
def f_eval(f_str, x):
    return eval(f_str, {"x": x, "math": math, **math.__dict__})

# ---------------------------------------------
# Cuadratura de Gauss-Legendre general
# ---------------------------------------------
def gauss_legendre(f_str, a, b, n):
    # Obtiene nodos y pesos
    x, w = np.polynomial.legendre.leggauss(n)

    # Cambio de intervalo [-1,1] -> [a,b]
    t = 0.5 * (x + 1) * (b - a) + a

    # Evaluación
    S = 0
    for i in range(n):
        S += w[i] * f_eval(f_str, t[i])

    return 0.5 * (b - a) * S


# ---------------------------------------------
# PROGRAMA PRINCIPAL
# ---------------------------------------------
print("\n=== CUADRATURA DE GAUSS-LEGENDRE ===\n")

f_str = input("Ingrese la función f(x): ")
a = float(input("Límite inferior a: "))
b = float(input("Límite superior b: "))
n = int(input("Número de puntos de Gauss (n): "))

resultado = gauss_legendre(f_str, a, b, n)

print("\nResultado de la integral ≈", resultado)
```
